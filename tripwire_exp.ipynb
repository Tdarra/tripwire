```python
# =========================
# TripWire Experiments Notebook
# Gemini vs Classic Models — Temp & Safety Grid
# =========================

# --- Optional installs (uncomment in Colab) ---
!pip install -q google-generativeai scikit-learn xgboost matplotlib tqdm

import os, time, random, math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.auto import tqdm

from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix

# Optional XGBoost
try:
    from xgboost import XGBClassifier
    HAS_XGB = True
except Exception:
    HAS_XGB = False
    print("[INFO] xgboost not available; will skip XGBoost baseline.")

# Gemini
import google.generativeai as genai

# -----------------
# Global config
# -----------------
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE); random.seed(RANDOM_STATE)

MODEL_NAME = "gemini-1.5-flash"
TEMPS = [round(x,1) for x in np.arange(0.0, 2.0 + 1e-9, 0.2)]  # 0.0..2.0 step 0.2
FIXED_TEMP_FOR_SAFETY = 0.2
MAX_OUTPUT_TOKENS = 8

# File paths (relative to notebook cwd)
TRAIN_5_PATH = "tripwire_train_5pct_scam.csv"
TRAIN_20_PATH = "tripwire_train_20pct_scam.csv"
VAL_5_PATH = "tripwire_val_5pct_scam.csv"
VAL_20_PATH = "tripwire_val_20pct_scam.csv"


# -----------------
# Read Data
# -----------------

train_5 = pd.read_csv(TRAIN_5_PATH)
train_20 = pd.read_csv(TRAIN_20_PATH)
val_5 = pd.read_csv(VAL_5_PATH)
val_20 = pd.read_csv(VAL_20_PATH)


# Quick sanity
print("Train sizes:", len(train_5), len(train_20))
print("Val sizes:", len(val_5), len(val_20))
print("Overlap train5-val5:", len(set(train_5['text']).intersection(set(val_5['text']))))
print("Overlap train20-val20:", len(set(train_20['text']).intersection(set(val_20['text']))))

# -----------------
# Metrics helpers
# -----------------
def metrics_dict(y_true, y_pred, positive="SCAM"):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    to_bin = lambda a: (a == positive).astype(int)
    acc = accuracy_score(y_true, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(
        to_bin(y_true), to_bin(y_pred), average="binary", zero_division=0
    )
    return {"accuracy": acc, "precision": prec, "recall": rec, "f1": f1}

def print_metrics(title, m):
    print(title)
    for k in ["accuracy", "precision", "recall", "f1"]:
        print(f"  {k}: {m[k]:.4f}")

# -----------------
# Classic baselines
# -----------------
def run_classic(train_df, val_df, use_xgb=HAS_XGB):
    Xtr, ytr = train_df["text"].values, train_df["label"].values
    Xte, yte = val_df["text"].values, val_df["label"].values

    results = {}

    # Logistic Regression (string labels OK)
    logreg = Pipeline([
        ("tfidf", TfidfVectorizer(ngram_range=(1,2), max_features=20000)),
        ("clf", LogisticRegression(max_iter=1000, class_weight="balanced", solver="liblinear")),
    ])
    logreg.fit(Xtr, ytr)
    yhat_lr = logreg.predict(Xte)
    results["LogReg"] = metrics_dict(yte, yhat_lr)

    # XGBoost needs numeric labels
    if use_xgb:
        pos = (ytr == "SCAM").sum()
        neg = (ytr == "SAFE").sum()
        spw = float(neg / max(pos, 1))

        xgb = Pipeline([
            ("tfidf", TfidfVectorizer(ngram_range=(1,2), max_features=30000)),
            ("clf", XGBClassifier(
                n_estimators=500,
                max_depth=6,
                learning_rate=0.08,
                subsample=0.9,
                colsample_bytree=0.9,
                objective="binary:logistic",
                eval_metric="logloss",
                scale_pos_weight=spw,
                tree_method="hist",
                n_jobs=2,
                random_state=RANDOM_STATE
            )),
        ])

        # Train with 0/1 target
        ytr_bin = (ytr == "SCAM").astype(int)
        xgb.fit(Xtr, ytr_bin)

        # Predict probabilities, then threshold to return SAFE/SCAM
        proba = xgb.predict_proba(Xte)[:, 1]
        yhat_xgb = np.where(proba >= 0.5, "SCAM", "SAFE")

        results["XGBoost"] = metrics_dict(yte, yhat_xgb)

    return results

classic_5 = run_classic(train_5, val_5)
classic_20 = run_classic(train_20, val_20)
print("== Classic baselines on 5% SCAM validation ==")
for k,v in classic_5.items(): print_metrics(k, v)
print("\n== Classic baselines on 20% SCAM validation ==")
for k,v in classic_20.items(): print_metrics(k, v)

# -----------------
# Gemini setup
# -----------------
import getpass
if not os.environ.get("GEMINI_API_KEY"):
    os.environ["GEMINI_API_KEY"] = getpass.getpass("Enter GEMINI_API_KEY (not echoed): ")
genai.configure(api_key=os.environ["GEMINI_API_KEY"])

def build_prompt(msg: str) -> str:
    return f"""You are a strict binary classifier for consumer scam detection.

Rules (examples of likely SCAM):
- Unsolicited money requests, gift cards, crypto, urgent payment or account lock warnings.
- Claims of prizes, refunds, tax rebates requiring action or links.
- Impersonation of banks, delivery, government, merchants; suspicious shortened links.
- Poor grammar + pressure tactics.

Output contract:
- Return exactly one word: SCAM or SAFE.
- No punctuation, no quotes, no explanations.

Classify the following message:
\"{msg.strip()}\""""

def normalize_label(text: str) -> str:
    if not text:
        return "UNCERTAIN"
    t = text.strip().upper()
    if t.startswith("SCAM"): return "SCAM"
    if t.startswith("SAFE"): return "SAFE"
    return "UNCERTAIN"

def classify_with_gemini(texts, temperature=0.0, safety_settings=None, sleep_sec=0.0, retries=3, backoff=1.5):
    model = genai.GenerativeModel(
        model_name=MODEL_NAME,
        generation_config={
            "temperature": float(temperature),
            "top_k": 1,
            "top_p": 0.0,
            "max_output_tokens": MAX_OUTPUT_TOKENS
        },
        safety_settings=safety_settings or [
            {"category": "HARM_CATEGORY_DANGEROUS_CONTENT", "threshold": "BLOCK_ONLY_HIGH"},
            {"category": "HARM_CATEGORY_HATE_SPEECH",       "threshold": "BLOCK_ONLY_HIGH"},
            {"category": "HARM_CATEGORY_HARASSMENT",        "threshold": "BLOCK_ONLY_HIGH"},
            {"category": "HARM_CATEGORY_SEXUALLY_EXPLICIT", "threshold": "BLOCK_ONLY_HIGH"},
        ]
    )
    preds = []
    for msg in tqdm(texts, leave=False):
        attempt = 0
        while True:
            try:
                resp = model.generate_content(build_prompt(msg))
                text = getattr(resp, "text", None) or ""
                preds.append(normalize_label(text))
                break
            except Exception as e:
                attempt += 1
                if attempt > retries:
                    print(f"[WARN] giving up after {retries} retries: {e}")
                    preds.append("UNCERTAIN")
                    break
                time.sleep(backoff ** attempt)
        if sleep_sec > 0:
            time.sleep(sleep_sec)
    return preds

def evaluate_preds(y_true, y_pred):
    m = metrics_dict(y_true, y_pred)
    return {k: float(m[k]) for k in m}

# -----------------
# Experiment 1 — Temperature sweep (0.0 → 2.0)
# -----------------
results_temp = {"5% SCAM": [], "20% SCAM": []}

for t in TEMPS:
    print(f"== Temperature {t} ==")
    preds_5 = classify_with_gemini(val_5["text"].tolist(), temperature=t)
    m5 = evaluate_preds(val_5["label"].tolist(), preds_5); m5["temperature"] = t
    results_temp["5% SCAM"].append(m5)

    preds_20 = classify_with_gemini(val_20["text"].tolist(), temperature=t)
    m20 = evaluate_preds(val_20["label"].tolist(), preds_20); m20["temperature"] = t
    results_temp["20% SCAM"].append(m20)

df_temp_5 = pd.DataFrame(results_temp["5% SCAM"]).sort_values("temperature").reset_index(drop=True)
df_temp_20 = pd.DataFrame(results_temp["20% SCAM"]).sort_values("temperature").reset_index(drop=True)

print("\n== Temperature sweep metrics (5% SCAM) ==")
display(df_temp_5)
print("\n== Temperature sweep metrics (20% SCAM) ==")
display(df_temp_20)

def plot_temp_curve(df_curve, classic_metrics, title):
    plt.figure(figsize=(8,5))
    plt.plot(df_curve["temperature"], df_curve["f1"], marker="o", label="Gemini F1")
    plt.plot(df_curve["temperature"], df_curve["accuracy"], marker="o", linestyle="--", label="Gemini Accuracy")
    for name, m in classic_metrics.items():
        plt.axhline(y=m["f1"], linestyle=":", label=f"{name} F1 (classic)")
    plt.xlabel("Temperature"); plt.ylabel("Score"); plt.ylim(0, 1.05)
    plt.grid(True, alpha=0.3); plt.title(title); plt.legend(); plt.show()

plot_temp_curve(df_temp_5, classic_5, "5% SCAM — Gemini vs Classic baselines")
plot_temp_curve(df_temp_20, classic_20, "20% SCAM — Gemini vs Classic baselines")

# -----------------
# Experiment 2 — Safety settings grid (fixed temperature = 0.2)
# -----------------
def safety_all(threshold):
    return [
        {"category": "HARM_CATEGORY_DANGEROUS_CONTENT", "threshold": threshold},
        {"category": "HARM_CATEGORY_HATE_SPEECH",       "threshold": threshold},
        {"category": "HARM_CATEGORY_HARASSMENT",        "threshold": threshold},
        {"category": "HARM_CATEGORY_SEXUALLY_EXPLICIT", "threshold": threshold},
    ]

SAFETY_CONFIGS = {
    "default_only_high": safety_all("BLOCK_ONLY_HIGH"),
    "all_none":          safety_all("BLOCK_NONE"),
    "all_medium_plus":   safety_all("BLOCK_MEDIUM_AND_ABOVE"),
    "all_low_plus":      safety_all("BLOCK_LOW_AND_ABOVE"),
    # Per-category ablations (others at ONLY_HIGH)
    "none_danger_only_high_others": [
        {"category": "HARM_CATEGORY_DANGEROUS_CONTENT", "threshold": "BLOCK_NONE"},
        {"category": "HARM_CATEGORY_HATE_SPEECH",       "threshold": "BLOCK_ONLY_HIGH"},
        {"category": "HARM_CATEGORY_HARASSMENT",        "threshold": "BLOCK_ONLY_HIGH"},
        {"category": "HARM_CATEGORY_SEXUALLY_EXPLICIT", "threshold": "BLOCK_ONLY_HIGH"},
    ],
    "none_hate_only_high_others": [
        {"category": "HARM_CATEGORY_DANGEROUS_CONTENT", "threshold": "BLOCK_ONLY_HIGH"},
        {"category": "HARM_CATEGORY_HATE_SPEECH",       "threshold": "BLOCK_NONE"},
        {"category": "HARM_CATEGORY_HARASSMENT",        "threshold": "BLOCK_ONLY_HIGH"},
        {"category": "HARM_CATEGORY_SEXUALLY_EXPLICIT", "threshold": "BLOCK_ONLY_HIGH"},
    ],
    "none_harass_only_high_others": [
        {"category": "HARM_CATEGORY_DANGEROUS_CONTENT", "threshold": "BLOCK_ONLY_HIGH"},
        {"category": "HARM_CATEGORY_HATE_SPEECH",       "threshold": "BLOCK_ONLY_HIGH"},
        {"category": "HARM_CATEGORY_HARASSMENT",        "threshold": "BLOCK_NONE"},
        {"category": "HARM_CATEGORY_SEXUALLY_EXPLICIT", "threshold": "BLOCK_ONLY_HIGH"},
    ],
    "none_sexual_only_high_others": [
        {"category": "HARM_CATEGORY_DANGEROUS_CONTENT", "threshold": "BLOCK_ONLY_HIGH"},
        {"category": "HARM_CATEGORY_HATE_SPEECH",       "threshold": "BLOCK_ONLY_HIGH"},
        {"category": "HARM_CATEGORY_HARASSMENT",        "threshold": "BLOCK_ONLY_HIGH"},
        {"category": "HARM_CATEGORY_SEXUALLY_EXPLICIT", "threshold": "BLOCK_NONE"},
    ],
}

def run_safety_grid(val_df, temp=FIXED_TEMP_FOR_SAFETY):
    results = []
    texts = val_df["text"].tolist(); labels = val_df["label"].tolist()
    for name, cfg in SAFETY_CONFIGS.items():
        print(f"Config: {name}")
        preds = classify_with_gemini(texts, temperature=temp, safety_settings=cfg)
        m = evaluate_preds(labels, preds); m["config"] = name
        results.append(m)
    return pd.DataFrame(results)

df_safety_5 = run_safety_grid(val_5)
df_safety_20 = run_safety_grid(val_20)

print("\n== Safety grid (5% SCAM) =="); display(df_safety_5)
print("\n== Safety grid (20% SCAM) =="); display(df_safety_20)

def plot_safety(df, title):
    order = list(df.sort_values("f1", ascending=False)["config"])
    d = df.set_index("config").loc[order]
    plt.figure(figsize=(10,5))
    plt.bar(range(len(order)), d["f1"].values)
    plt.xticks(range(len(order)), order, rotation=30, ha="right")
    plt.ylim(0, 1.05); plt.ylabel("F1"); plt.title(title)
    plt.grid(True, axis="y", alpha=0.3)
    plt.show()

plot_safety(df_safety_5, "Safety configs @ T=0.2 (5% SCAM)")
plot_safety(df_safety_20, "Safety configs @ T=0.2 (20% SCAM)")
```
